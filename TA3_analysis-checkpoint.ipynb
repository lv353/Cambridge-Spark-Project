### My PhD focuses on investigating exciton dynamics in hybrid systems comprised of lanthanide doped nanoparticles coupled to organic semiconductors.

I will try to explain the basics of what that sentence means here: When materials in nature are undisturbed, they are in their ground state, which means that they are in their lowest energy configuration. All materials contain a certain number of electrons and these are all placed in molecular orbitals. The important thing to know about these molecular orbitals is that each can accommodate 2 electrons only (in the Figure below, the horizontal lines represent molecular orbitals). Electrons in general have an intrinsic property which is called their spin, and this spin can have two different values that we will call 'spin-up' (+ 1/2 usually denoted) and 'spin-down' (-1/2 usually denoted). The arrows in the Figure below represent this spin property.

Most materials in their ground state are so-called **singlet materials**; this means that the net overall spin number (denoted S in the Figure below) is 0. To make this a bit more concrete: The singlet ground state in the Figure below on the left can be seen to have an electron with spin-up and one with spin-down. If we compute the overall spin number S by adding the value of a spin-up electron (+ 1/2) and a spin-down electron (- 1/2) we find an overall spin number of zero. The term singlet for such systems comes from their multiplicity M, which can be calculated using M = 2S + 1. For S = 0, a multiplicity value of M is 1, corresponding to a singlet, is found.

When light is now shined onto the material, energy can be absorbed by the electrons which will lead, as shown below, to an electron being promoted to a higher lying molecular orbital. This will leave behind a so-called **hole** (also called an electron-hole as it refers to the absence of an electron). These new electronic states that can be created this way are so-called excited states. As before, if the overall spin number is 0, we will be left with a **singlet excited state**, as shown in the middle in the Figure below. If however, a spin flip of the promoted electron occurs, the overal spin number S will now become 1 (2 * +1/2) and a multiciplity M of 3 will result, which is referred to as an **triplet excited state**. In both the case of a singlet and triplet excited state, the pair consisting of the excited electron and the hole is referred to as an **exciton** and these **excitons are known to be able to move through a material**. 

<img src="triplet_states.png" style="width: 605px;"/>

The excitons described above will after a certain time lose their 'excess energy' and return to their ground state. **The main goal of my work is to get an insight into how rapidly this happens, and also into how far these excitons can move through a material so that their energy can be extracted elsewhere.** 

In order to study the motion and dynamics of these so-called **excitons**, I use various ultrafast spectroscopy techniques with time resolution that are referred to as **transient absorption spectroscopy**. Transient absorption spectroscopy is a pump-probe spectroscopic technique used to measure the absorption and lifetimes of excited states of molecules and materials. The pump-probe technique is illustrated below and can be described by a sample being excited by a light source (pump pulse), after which a second light source (probe pulse) is sent in to measure changes in absorption as a function of wavelength and time.

My work utilises this technique to study singlet and triplet exciton transport in organic semiconductors, in nanoparticles that contain lanthanide ions, and in hybrid systems that are comprised of organic semiconductors to these nanoparticles. The example dataset used in here shows the transient absorption data obtained in a measurement on an organic semiconductor only.

The **spectral background** (which is theoretically speaking the signal average before time zero) needs to be **removed**. Below the time can be entered up until which the background signal will be calculated (i.e. signal calculated from start time till *time_background_removal*)

In the cells below, a dataframe will be created containing the data that will be used to determine the average spectral background.

```python
def list_background_values(dataframe):
    
    time_background_mask = dataframe[0] == time_background_removal              
    row_number_time_background_removal = dataframe.index[time_background_mask].tolist()

    # Create dataframe of data to be averaged
    df_background_to_be_averaged = dataframe.copy()

    df_background_to_be_averaged = df_background_to_be_averaged.loc[0:row_number_time_background_removal[0],:]

    # Drop row and column 0
    df_background_to_be_averaged = df_background_to_be_averaged.drop(0)
    df_background_to_be_averaged = df_background_to_be_averaged.drop(0,axis=1)
    df_background_to_be_averaged

    # Create list with background values at each wavelength
    list_with_average_to_be_subtracted_values = df_background_to_be_averaged.mean(axis=0).tolist()
    
    return df_background_to_be_averaged.mean(axis=0).tolist()
```

```python
# Create list with background values at each wavelength
list_with_average_to_be_subtracted_values = list_background_values(df_average)
```

The cell below contain functions that takes in the average dataframe before background subtraction and **returns a dataframe in which the background is removed**. The first function returns a dataframe without the wavelength and time column included, the second function does include the wavelength and time column.

```python
def create_updated_average_df(average_df, to_be_subtracted_values):
    
    average_df = average_df.copy()
    average_df = average_df.T    
    
    for column, value in zip(average_df, range(len(to_be_subtracted_values))):
        average_df[column] = average_df[column] - to_be_subtracted_values[value]
    
    average_df = average_df.T
    
    return average_df

updated_data = create_updated_average_df(data_average, list_with_average_to_be_subtracted_values)
```

```python
def create_updated_average_df_with_time_and_wavelength(updated_df, times, wavelengths):
    
    full_background_subtracted_average_frame = updated_df.copy()
    full_background_subtracted_average_frame.insert(loc=0, column=0, value=wavelengths)
    full_background_subtracted_average_frame = full_background_subtracted_average_frame.T
    new_time_average = times.copy()
    new_time_average.insert(0,0)
    full_background_subtracted_average_frame.insert(loc=0, column=0, value=new_time_average)

    return full_background_subtracted_average_frame
```

Below five functions are presented that will be used later on and can be used, respectively, to:
- **Set a new time zero**, i.e. shift the time data horizontally in the time domain.
- **Define evenly spread line colors** in a viridis colourmap.
- Create a dataframe of time versus delta transmittance values that are averaged over the wavelength; this is referred to as a **kinetic traces dataframe**.
- **Determine the average value of a list of values**.
- **Determine percentage difference of values in a list in comparison to a certain value**.

```python
def time_to_new_time(time,new_t0):
    return time-new_t0
```

```python
def figure_colors(list_of_ranges):
    evenly_spaced_interval = np.linspace(0, 1, len(list_of_ranges))
    return [cm.viridis(x) for x in evenly_spaced_interval]
```

```python
def kinetic_traces_df(dataframe_kinetics, wavelength_range):
        
    starting_column = wavelength_average.index(closest(wavelength_average, wavelength_range[0]))
    final_column = wavelength_average.index(closest(wavelength_average, wavelength_range[1]))

    dataframe_loop = dataframe_kinetics.copy()
    dataframe_loop.drop(dataframe_loop.columns[final_column:], axis=1, inplace=True)
    dataframe_loop.drop(dataframe_loop.columns[1:starting_column], axis=1, inplace=True)

    kinetic_traces_frame = pd.DataFrame({'time': dataframe_loop.loc[1:,0].tolist()})

    dataframe_loop.drop(0, axis=0, inplace=True)
    dataframe_loop = dataframe_loop.T
    dataframe_loop.drop(0, axis=0, inplace=True)

    list_of_means = []

    for column in dataframe_loop:
        list_of_means.append(dataframe_loop[column].mean())

    kinetic_traces_frame['average'] = list_of_means
    kinetic_traces_frame['time'] = time_to_new_time(kinetic_traces_frame['time'],-2.5)
                        
    return kinetic_traces_frame
```

```python
def average(lst): 
        return sum(lst) / len(lst)
```

```python
def percentage_difference(lst,mean):
    
    percentage_difference_list = []

    for entry in lst:
        value = abs((entry - mean) / mean) * 100.0
        percentage_difference_list.append(value)
    
    return percentage_difference_list
```

```python
def lin_scale_plot(data, time_list, wavelength_list, plot_type, wavelength_range, time_range):
    
    if plot_type == 'contour':
        
        time, wavelength = np.meshgrid(time_list, wavelength_list)
        
        plt.figure(
            figsize=(7,5)
        )
        
        plt.contourf(
            time, 
            wavelength, 
            data,
            cmap="viridis"
        );

        plt.colorbar();
    
        plt.ylim(wavelength_range)
        plt.xlabel('time (ns)')
        plt.ylabel('wavelength (nm)')
        
        return plt.show()
    
    elif plot_type == 'spectra':
        
        wavelength_start_index = wavelength_list.index(closest(wavelength_list, wavelength_range[0]))
        wavelength_end_index = wavelength_list.index(closest(wavelength_list, wavelength_range[1]))
        
        colors = figure_colors(time_range)
        colors_index = -1
        
        plt.axhline(
            y=0,
            color='black'
        )

        for i in range(len(time_range)):

            starting_column = time_list.index(closest(time_list, time_range[i][0]))
            final_column = time_list.index(closest(time_list, time_range[i][1]))

            colors_index += 1

            dataframe_loop = data.T.copy()

            dataframe_loop.drop(dataframe_loop.columns[final_column+2:], axis=1, inplace=True)
            dataframe_loop.drop(dataframe_loop.columns[1:starting_column+1], axis=1, inplace=True)

            spectra_frame = pd.DataFrame({'wavelength': dataframe_loop.loc[1:,0].tolist()[wavelength_start_index-1:wavelength_end_index+2]})

            dataframe_loop.drop(0, axis=0, inplace=True)

            dataframe_loop = dataframe_loop.T

            dataframe_loop.drop(0, axis=0, inplace=True)
            list_of_means = []

            dataframe_loop.drop(dataframe_loop.columns[wavelength_end_index+2:], axis=1, inplace=True)
            dataframe_loop.drop(dataframe_loop.columns[:wavelength_start_index-1], axis=1, inplace=True)

            for column in dataframe_loop:
                list_of_means.append(dataframe_loop[column].mean())

            spectra_frame['dT_values'] = list_of_means
  
            df_label = f'{str(time_range[i][0])} - {str(time_range[i][1])} ns'

            plt.plot(
                spectra_frame['wavelength'],
                spectra_frame['dT_values'],
                label = df_label,
                color = colors[colors_index]
            )

            plt.legend(
                frameon=False,
                #title='range', 
                #bbox_to_anchor=(1,1), 
                #loc='upper left', 
                #prop={"size":12}
            )

        #plt.xlim([0.1,time[-1]])
        #plt.setp(axLin.get_xticklabels(), visible=True);
        plt.xlim((wavelength_range[0], wavelength_range[1]))
        plt.xlabel('wavelength (nm)')
        plt.ylabel('dT/T')
        return plt.show()
```

The cell below tries to identify **possible outliers** in the sweeps graph above. Visible inspection will, however, remain necessary to confirm whether the possible detected outlier sweeps are to be removed.

```python
# Calculate total average of each dataframe in list of sweeps for certain wavelength range

# Set threshold above which potential outlier are defined
threshold = 10

list_of_overall_means = []
list_of_average_means = []

for dataframe in list_of_sweep_dataframes:
    
    list_of_average_means = []
    
    average_sweep_trial = dataframe.copy()
    
    average_sweep_trial.drop(average_sweep_trial.columns[final_column:], axis=1, inplace=True)
    average_sweep_trial.drop(average_sweep_trial.columns[:starting_column], axis=1, inplace=True)
    
    average_sweep_trial.drop(0,inplace=True)
    
    for column in average_sweep_trial:
        list_of_average_means.append(average_sweep_trial[column].mean())
    
    list_of_overall_means.append(average(list_of_average_means))

mean = sum(list_of_overall_means)/len(list_of_overall_means)

for entry in range(len(percentage_difference(list_of_overall_means, mean))):
    if percentage_difference(list_of_overall_means, mean)[entry] > threshold:
        print(f"Potential outlier in sweep {entry+1}.")
```

## This part implements basic PCA analysis to the collected data both in the wavelength and time domain.

Having visualised the data and having a tool that can present the data in its 3D format, as well as averaged over the wavelength or time domain, we move towards further analysis of the data.

Firstly, a tool is implemented to carry out principal component analysis (PCA). This is an unsupervisod machine learning method aimed at reducing the dimensionality of large data sets, which is done by transforming a large set of variables into a smaller one that still contains most of the information in the large set. In order to carry out this analysis, a relationship between all time columns in the wavelength domain or all wavelength columns in the time domain is sought after. Ideally, this analysis will give us a rough idea of the number of species present in the collected data.

Firstly, two **PCA dataframes are created for a specified wavelength and time domain**. For this, the desired wavelength range and time range must be added below as a tuple inside a list.

```python
# Enter wavelength range of the experiment
wavelength_range_list = [(300, 650)]
time_range_list = [(-5, 500)]

# The function below can be used to compute a PCA_wavelength and PCA_time dataframe
def PCA_dataframe(full_dataframe, wavelength_range, time_range, df_type):
    
    PCA_df = full_dataframe.copy()
    
    starting_wavelength_column = wavelength_average.index(closest(wavelength_average, wavelength_range[0][0]))
    final_wavelength_column = wavelength_average.index(closest(wavelength_average, wavelength_range[0][1]))
    starting_time_column = time_average.index(closest(time_average, time_range[0][0]))
    final_time_column = time_average.index(closest(time_average, time_range[0][1]))
    
    if df_type == 'wavelength':
        PCA_df.drop(PCA_df.columns[final_time_column+1:], axis=1, inplace=True)
        PCA_df.drop(PCA_df.columns[:starting_time_column], axis=1, inplace=True)
        PCA_df['wavelength'] = wavelength_average
        PCA_df = PCA_df.T
        PCA_df.drop(PCA_df.columns[final_wavelength_column+1:], axis=1, inplace=True)
        PCA_df.drop(PCA_df.columns[:starting_wavelength_column], axis=1, inplace=True)
    elif df_type == 'time':
        PCA_df = PCA_df.T
        PCA_df.drop(PCA_df.columns[final_wavelength_column+1:], axis=1, inplace=True)
        PCA_df.drop(PCA_df.columns[:starting_wavelength_column], axis=1, inplace=True)
        PCA_df['time'] = time_average
        PCA_df = PCA_df.T
        PCA_df.drop(PCA_df.columns[final_time_column+1:], axis=1, inplace=True)
        PCA_df.drop(PCA_df.columns[:starting_time_column], axis=1, inplace=True)

    return PCA_df.T

PCA_wavelength_df = PCA_dataframe(updated_data, wavelength_range_list, time_range_list, 'wavelength')
PCA_time_df =  PCA_dataframe(updated_data, wavelength_range_list, time_range_list, 'time')
```

The following function creates a **specific PCA dataframe in the wavelength and time domain that will be used for the actual PCA analysis**. It also separates and saves the either wavelength or time list corresponding to the previously defined range.

```python
def PCA_analysis_df(dataframe_for_PCA, df_type):

    PCA_data_analysis = dataframe_for_PCA.copy()
    
    if df_type == 'wavelength':
        wavelength_list = PCA_data_analysis[df_type]
        wavelength_list = wavelength_list.reset_index()
        wavelength_list.drop('index', axis=1,inplace=True)
        PCA_data_analysis.drop(df_type, axis=1, inplace=True)
        return PCA_data_analysis, wavelength_list[df_type].tolist()
    
    elif df_type == 'time':
        time_list = PCA_data_analysis[df_type]
        time_list = time_list.reset_index()
        time_list.drop('index', axis=1,inplace=True)
        PCA_data_analysis.drop(df_type, axis=1, inplace=True)
        return PCA_data_analysis, time_list[df_type].tolist()
    
PCA_analysis_wavelength_df, PCA_analysis_wavelength_list = PCA_analysis_df(PCA_wavelength_df, 'wavelength')
PCA_analysis_time_df, PCA_analysis_time_list = PCA_analysis_df(PCA_time_df, 'time')
```

## This section focuses on fitting the kinetic traces obtained in the transient absorption spectroscopy measurements.

The kinetic traces that were visualised earlier for specific wavelength regions seem to correspond to separate species, both confirmed by visual inspection of the obtained spectra, as well as by the PCA analysis showing the presence of most likely 2 or 3 distinct species. 

**Fitting of the decay of these kinetic traces is generally done using exponential decay functions** and is extremely important as it gives direct information about the lifetime of the inspected species. The **initial rise of the signal** is usually described either by a **gaussian rise function** if the rise is limited by the instrument response function (which basically gives the minimal time resolution of your detector) or by an **exponential rise** if the rise time is larger than the instrument response function.

Below a fitting scheme has been set up that can be used to fit extracted kinetic traces. For this purpose, **a wavelength range must be entered first for the region that one wants to perform the fit in**. A dataframe is then created that contains the information for the desired kinetic trace, of which a quick plot on a logarithmic scale is provided below.

```python
# The wavelength range for which the fit will be performed must be entered below as a tuple
range_list = [(320, 350), (440, 460)]

# The next bit of code creates a full dataframe of background subtracted data
full_background_subtracted_average_frame = create_updated_average_df_with_time_and_wavelength(updated_data, time_average, wavelength_average)

kinetic_traces_frame_list = []
x_total_list = []
y_total_list = []

for i in range(len(range_list)):
    kinetic_traces_frame_list.append(kinetic_traces_df(full_background_subtracted_average_frame, range_list[i]))


for i, frame in zip(range(len(kinetic_traces_frame_list)), kinetic_traces_frame_list):
    x_total_list.append(np.array(frame['time'][:].tolist()))
    y_total_list.append(np.array(frame['average'][:].tolist()))

    fig = plt.plot(
        x_total_list[i], 
        y_total_list[i]
    )

    plt.xscale('log')
    plt.xlabel('time (ns)')
    plt.ylabel('dT/T')
    plt.xlim((0.1,500))

    plt.show()
```

Below three functions are provided to **fit**, respectively:
- **A monoexponential decay with a y offset**
- **A Gaussian rise with a y offset**
- **A convoluted function of the Gaussian rise and monoexponential decay** (this function is not functional yet)

More functions need to be added to fit higher order exponential decays, but for this dataset a monoexponential decay is sufficient.

```python
# Function to calculate the exponential with constants a and b
def exponential_decay(time, a, b, y0):
    return a*np.exp(-b*time) + y0

def exponential_rise(time, a, b, y0):
    return a*np.exp(b*time) + y0

def gaussian_rise(time, a, sigma, mu, y1):
    return a*np.exp(-0.5*(time-mu)**2 / sigma**2) + y1

def convoluted_function_gaussian(time, a, b, sigma, mu, y0, y1):
    rise = a*np.exp(-0.5*(time-mu)**2/sigma**2) + y1
    decay = a*np.exp(b*time) + y0
    return signal.convolve(rise, decay, 'same') #/ sum(rise)

def convoluted_function_exponential(time, a, b, c, y0, y1):
    rise = a*np.exp(b*time) + y0
    decay = a*np.exp(-c*time) + y1
    return signal.convolve(rise, decay, 'same') / sum(rise)
```

Below **numpy arrays are created for the rising and decaying part of the kinetic traces of interest**. Firstly, the index of the minimum of the kinetic trace of interest is found (change this to max if the trace has the opposite shape)

```python
index_list = []
x_rise_list = []
x_decay_list = []
y_rise_list = []
y_decay_list = []

for kinetic_traces_frame in kinetic_traces_frame_list:
    index_list.append(kinetic_traces_frame['average'].tolist().index(kinetic_traces_frame['average'].min()))
 
for i, kinetic_traces_frame in zip(range(len(index_list)), kinetic_traces_frame_list):
    x_rise_list.append(np.array(kinetic_traces_frame['time'][:index_list[i]+1].tolist()))
    x_decay_list.append(np.array(kinetic_traces_frame['time'][index_list[i]:].tolist()))

    y_rise_list.append(np.array(kinetic_traces_frame['average'][:index_list[i]+1].tolist()))
    y_decay_list.append(np.array(kinetic_traces_frame['average'][index_list[i]:].tolist()))
```

In the next cell the **actual fitting of the rising and decaying part** is performed using the **curve_fit** method. The parameters for each of the fits are printed out below. A Gaussian rise is used here for the first fit since the rise time of the trace coincides with the instrument response function. An exponential rise is used for the second fit since the rise time is larger than the instrument response function

At the bottom of the cell, **the rise and decay fit are plotted** as dashed lines on top of the original data. Visual inspection shows an excellent agreement between the data and the fits.

```python
# Define a rise list, indicating if signal rise of the previously specified ranges is 'Gaussian' or 'exponential'
rise_list = ['Gaussian', 'exponential']

for i, rise in zip(range(len(rise_list)), rise_list):

    if rise == 'exponential':
        rise_pars, rise_cov = curve_fit(
            f=exponential_rise, 
            xdata=x_rise_list[i], 
            ydata=y_rise_list[i], 
            p0=[1, 1, 0], 
            bounds=(-np.inf, np.inf)
        )
    elif rise == 'Gaussian':
        rise_pars, rise_cov = curve_fit(
            f=gaussian_rise, 
            xdata=x_rise_list[i], 
            ydata=y_rise_list[i], 
            p0=[-1, -1, 1, 1], 
            bounds=(-np.inf, np.inf)
        )

    decay_pars, decay_cov = curve_fit(
        f=exponential_decay,
        xdata=x_decay_list[i],
        ydata=y_decay_list[i],
        p0=[-1, 1, 0],
        bounds=(-np.inf, np.inf)
    )

    # Get the standard deviations of the parameters (square roots of the # diagonal of the covariance) and residuals
    if rise == 'Gaussian':
        stdevs_rise = np.sqrt(np.diag(rise_cov))
        res_rise = y_rise_list[i] - gaussian_rise(x_rise_list[i], *rise_pars)
    if rise == 'exponential':
        stdevs_rise = np.sqrt(np.diag(rise_cov))
        res_rise = y_rise_list[i] - exponential_rise(x_rise_list[i], *rise_pars)

    stdevs_decay = np.sqrt(np.diag(decay_cov))
    res_decay = y_decay_list[i] - exponential_decay(x_decay_list[i], *decay_pars)

    if rise == 'Gaussian':
        print(f"The parameters for the Gaussian rise are: a = {rise_pars[0]}, sigma = {rise_pars[1]}, mu = {rise_pars[2]}, y1 = {rise_pars[3]}")
        #print(f"The covariance of the parameters fitted for the Gaussian rise are: {rise_cov}")
    elif rise == 'exponential':
        print(f"The parameters for the exponential rise are: a = {rise_pars[0]}, b = {rise_pars[1]}, y0 = {rise_pars[2]}")
        #print(f"The covariance of the parameters fitted for the Gaussian rise are: {rise_cov}")

    print(f"The parameters for the exponential decay are: a = {decay_pars[0]}, b = {decay_pars[1]}, y0 = {decay_pars[2]}")
    #print(f"The covariance of the parameters fitted for the exponential decay are: {decay_cov}")

    # This plots the original recorded data (i.e. kinetic trace)
    fig = plt.plot(
        x_total_list[i], 
        y_total_list[i], 
        color = 'blue'
    )

    # Plots the Gaussian rise fit
    if rise == 'Gaussian':
        plt.plot(
            x_rise_list[i], 
            gaussian_rise(x_rise_list[i], *rise_pars), 
            linestyle='--', 
            linewidth=1.5, 
            color='black'
        )

    # Plots an exponential rise
    elif rise == 'exponential':
        plt.plot(
            x_rise_list[i], 
            exponential_rise(x_rise_list[i], *rise_pars), 
            linestyle='--', 
            linewidth=1.5, 
            color='black'
        )

    # Plots the exponential decay fit
    plt.plot(
        x_decay_list[i],
        exponential_decay(x_decay_list[i], *decay_pars),
        linestyle='--',
        linewidth=1.5,
        color='black'
    )

    plt.xscale('log')
    plt.xlim((0.1,500))

    plt.show()
```

As can be seen from the fit, **this fit is well optimised for a convolution of an exponential rise followed by an exponential decay, but is not yet optimised for a Gaussian rise followed by an exponential decay**.

## This section implements a different method to try and fit the presented convoluted signal comprised of a Gaussian rise and exponential decay using the scipy optimise and minimise method

To illustrate this, the x and y data from the total data list are extracted for the frame that is characterised by a Gaussian rise and exponential decay.

```python
y_total = y_total_list[0]
x_total = x_total_list[0]
```

Here an **error function** is defined.

```python
def error_function(parameters, paramArgs):
    
    y_total = paramArgs[0]
    x_total = paramArgs[1]
    
    (A, b, y0, k1, mu, sigma) = parameters   
    
    yPred = []
    
    for i in range(y_total.shape[0]):
        yPred.append(conv_function(x_total[i], A, b, y0, k1, mu, sigma))   
        
    yPred = np.asarray(yPred)
    error = abs(yPred-y_total)**2  
    return np.sum(error)
```

Next, the actual **onvoluted function** is presented.

```python
def conv_function(x, A, b, y0, k1, mu, sigma):
    y1 = np.asarray(A*np.exp(b*x)) + y0
    y2 = np.asarray(A*np.exp(-0.5*(x-mu)**2/sigma**2)) + 0*k1  
    return np.convolve(y1, y2, mode = 'valid')[0]
```

I tried to add k1 as a parameter (y-offset) but ended up not using it since it yielded worst results.

```python
parameters = [-1, -1, 0.0001, 0.0001, 1, -1]

paramArgs = [y_total, x_total]        
(A, b, y0, k1, mu, sigma) = parameters  
```

This function **minimises the residuals** of the performed fit.

```python
res = minimize(error_function, parameters,paramArgs, method = 'BFGS', options={'gtol': 1e-08})
```

```python
optimal_params = res['x']
(A, b, y0, k1, mu, sigma) = optimal_params
```

```python
yPred = []

for i in range(y_total.shape[0]):
    yPred.append(conv_function(x_total[i], A, b, y0, k1, mu, sigma))   
    
yPred_optimal = np.asarray(yPred)
```

The obtained fit is in closer agreement with the data than the previous fit. However, the fit is not yet accurate and further optimisation of the model is necessary.

```python
fig = plt.plot(
    x_total, 
    y_total,
    color='blue'
)
plt.plot(
    x_total, 
    yPred_optimal, 
    linestyle='--', 
    linewidth=1.5, 
    color='black'
)

plt.xscale('log')

plt.xlim((0.1,500))

plt.show()
```

**Sequential differential imaging**: Divides the i^th frame by the (i-n)^th frame, and then subtracts 1 (to centre the scale on zero). These images then represent the fractional intensity changes over a period of n frames (*typically we use n=1, but we have found that the biphasic transitions or other slow changes are sometimes easier to visualise with a larger value, e.g. n=4*).

```python
n = 1 #Use images spaced by n for differential imaging

#make array to hold result
seqdiffimage = np.zeros((len(imagestack[:,0,0])-n, len(imagestack[0,:,0]), len(imagestack[0,0,:]) ))

#in the for loop, use indices from the raw image. i.e. start at index equal to blocksize to make the first differntial frame
for RawFrame in np.arange(n,len(imagestack[:,0,0])):
    #RawFrame is the frame index of each frame in the raw image stack
    #First demoninator image is the first frame in the raw image stack
    Denominator = imagestack[RawFrame-n,:,:]
    #First numerator image is the n^th frame in the raw image stack 
    Numerator = imagestack[RawFrame,:,:]
    #divide the two blocks (excluding pixels that are not part of the particle):
    Diff = np.divide(Numerator,Denominator, where=maskframe0>0)
            
    #define the new index for the differntial image
    DiffFrame = RawFrame - n
    #Save the differential frame
    seqdiffimage[DiffFrame,:,:] = Diff

#Subtract 1 to centre the contrast scale around 0    
seqdiffimage = np.subtract(seqdiffimage,1,where=maskframe0>0)
#Trim times array to the correct size (miss off first n frames)
times = times[n:]
```
